```python
z = df.groupby(['region','smoker'])['charges'].mean().reset_index()
z 
```

```python
pivot_df = z.pivot(index='region', columns= 'smoker',values='charges')
pivot_df
```

```python
plt.figure(figsize=(50,50))
pivot_df.plot(kind='bar')
plt.title('Mean Values by Category and Subcategory')
plt.xlabel('Children')
plt.ylabel('Charges')
plt.xticks(rotation=45)
plt.legend(title='Subcategory',loc='upper right')
plt.show()
```
