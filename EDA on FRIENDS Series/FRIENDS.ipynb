```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x=df['Season'],data=df,palette='seismic')
plt.title('Episode Count According to Season')
plt.xlabel('Season')
plt.ylabel('Episode')
plt.grid()
```

```python
df.groupby('Season')['Duration'].sum()
```

```python
Season_duration = df.groupby('Season')['Duration'].mean().reset_index()
Season_duration
```

```python
Season_Star = df.groupby('Season')['Stars'].mean().reset_index()
Season_Star
```

```python
df.groupby('Season')['Votes'].sum()
```

```python
df.groupby('Season')['Votes'].mean()
```

```python
sns.barplot(x=Season_duration['Season'],y=Season_duration['Duration'])
plt.title('Duration of Different Season')
plt.xlabel('Season')
plt.ylabel('Total Duration')
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=Season_Star['Season'],y=Season_Star['Stars'],marker='o')
plt.xlabel('Season')
plt.ylabel('Stars')
plt.show()
```

```python
Director_star = df.groupby('Director')['Stars'].mean().reset_index()
Director_star
```

```python
Highest_director = df.groupby('Director')['Stars'].max().reset_index()
Highest_director
```

```python
plt.figure(figsize=(50,30))
sns.lineplot(x=Highest_director['Director'],y=Highest_director['Stars'],marker='o')
plt.xlabel('Director')
plt.ylabel('Stars')
plt.show()
```

```python
Lowest_Director = df.groupby('Director')['Stars'].min().reset_index()
Lowest_Director 
```

```python
plt.figure(figsize=(50,30))
sns.lineplot(x=Lowest_Director['Director'],y=Lowest_Director['Stars'],marker='o')
plt.xlabel('Director')
plt.ylabel('Stars')
plt.show()
```
